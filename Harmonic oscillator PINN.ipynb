###### 1D harmonic oscillator physics-informed neural network (PINN)

## Problem overview

The example problem we solve here is the 1D damped harmonic oscillator:
$$
m \dfrac{d^2 x}{d t^2} + \mu \dfrac{d x}{d t} + kx = 0~,
$$
with the initial conditions
$$
x(0) = 1~~,~~\dfrac{d x}{d t} = 0~.
$$
We will focus on solving the problem for the under-damped state, i.e. when 
$$
\delta < \omega_0~,~~~~~\mathrm{with}~~\delta = \dfrac{\mu}{2m}~,~\omega_0 = \sqrt{\dfrac{k}{m}}~.
$$
This has the following exact solution:
$$
x(t) = e^{-\delta t}(2 A \cos(\phi + \omega t))~,~~~~~\mathrm{with}~~\omega=\sqrt{\omega_0^2 - \delta^2}~.
$$

This problem was inspired by the following blog post: https://beltoforion.de/en/harmonic_oscillator/.

## Workflow overview

>First we will train a standard neural network to interpolate a small part of the solution, using some observed training points from the solution.

>Next, we will train a PINN to extrapolate the full solution outside of these training points by penalising the underlying differential equation in its loss function.

````python
from PIL import Image

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
````

````python
def save_gif_PIL(outfile, files, fps=5, loop=0):
    "Helper function for saving GIFs"
    imgs = [Image.open(file) for file in files]
    imgs[0].save(fp=outfile, format='GIF', append_images=imgs[1:], save_all=True, duration=int(1000/fps), loop=loop)
    
def oscillator(d, w0, x):
    """Defines the analytical solution to the 1D underdamped harmonic oscillator problem. 
    Equations taken from: https://beltoforion.de/en/harmonic_oscillator/"""
    assert d < w0
    w = np.sqrt(w0**2-d**2)
    phi = np.arctan(-d/w)
    A = 1/(2*np.cos(phi))
    cos = torch.cos(phi+w*x)
    sin = torch.sin(phi+w*x)
    exp = torch.exp(-d*x)
    y  = exp*2*A*cos # e^(-dt) * (2 * Acos(phi + wt) )
    return y

class FCN(nn.Module):
    "Defines a connected network"
    
    def __init__(self, N_INPUT, N_OUTPUT, N_HIDDEN, N_LAYERS):
        super().__init__()
        activation = nn.Tanh
        self.fcs = nn.Sequential(*[
                        nn.Linear(N_INPUT, N_HIDDEN),
                        activation()])
        self.fch = nn.Sequential(*[
                        nn.Sequential(*[
                            nn.Linear(N_HIDDEN, N_HIDDEN),
                            activation()]) for _ in range(N_LAYERS-1)])
        self.fce = nn.Linear(N_HIDDEN, N_OUTPUT)
        
    def forward(self, x):
        x = self.fcs(x)
        x = self.fch(x)
        x = self.fce(x)
        return x
````

## Generate training data

> First, we generate some training data from a small part of the true analytical solution.

For this problem, we use $\delta=2$, $\omega_0=20$, and try to learn the solution over the domain $x\in [0,1]$.

````python
d, w0 = 2, 20

# get the analytical solution over the full domain
x = torch.linspace(0,1,500).view(-1,1) # create nx1 arr of x values
y = oscillator(d, w0, x).view(-1,1) # create corresponding tensor with output values
print(x.shape, y.shape)

# slice out a small number of points from the LHS of the domain
x_data = x[0:200:20]
y_data = y[0:200:20]
print(x_data.shape, y_data.shape)

plt.figure()
plt.plot(x, y, label="Exact solution")
plt.scatter(x_data, y_data, color="tab:orange", label="Training data")
plt.legend()
plt.show()
````

## Normal neural network

> Next, we train a standard neural network (fully connected network) to fit these training points.

>We find that the network is able to fit the solution very closely in the vicinity of the training points, but does not learn an accurate solution outside of them.

````python
def plot_result(x,y,x_data,y_data,yh,xp=None, mu=None, k=None, mu_true=None, k_true=None):
    "Pretty plot training results"
    plt.figure(figsize=(8,4))
    plt.plot(x,y, color="grey", linewidth=2, alpha=0.8, label="Exact solution")
    plt.plot(x,yh, color="tab:blue", linewidth=4, alpha=0.8, label="Neural network prediction")
    plt.scatter(x_data, y_data, s=60, color="tab:orange", alpha=0.4, label='Training data')
    if xp is not None:
        plt.scatter(xp, -0*torch.ones_like(xp), s=60, color="tab:green", alpha=0.4, 
                    label='Physics loss training locations')
    l = plt.legend(loc=(1.01,0.34), frameon=False, fontsize="large")
    plt.setp(l.get_texts(), color="k")
    plt.xlim(-0.05, 1.05)
    plt.ylim(-1.1, 1.1)
    plt.text(1.065, 0.95, f"Training step: {i+1}", fontsize="xx-large", color="k")

    # hack fix: modify to display learned parameters
    if mu is not None and k is not None:
        mu_str = f"μ = {mu.item():.3f}"
        k_str = f"k = {k.item():.3f}"
        if mu_true is not None and k_true is not None:
            mu_str += f"  (true {mu_true:.3f})"
            k_str += f"  (true {k_true:.3f})"
        plt.text(1.065, 0.78, mu_str, fontsize="large", color="k")
        plt.text(1.065, 0.65, k_str, fontsize="large", color="k")
    
    plt.axis("off")
    
import os

# train standard neural network to fit training data
torch.manual_seed(123)
model = FCN(1,1,32,3)
optimizer = torch.optim.Adam(model.parameters(),lr=1e-3)
files = []
for i in range(1000):
    optimizer.zero_grad()
    yh = model(x_data)
    loss = torch.mean((yh-y_data)**2)# use mean squared error
    loss.backward()
    optimizer.step()
    
    
    # plot the result as training progresses
    if (i+1) % 10 == 0: 
        
        yh = model(x).detach()
        
        plot_result(x,y,x_data,y_data,yh)
        
        file = "plots/nn_%.8i.png"%(i+1)
        #path = os.path.join('plots', file)
        plt.savefig(file, format='png', bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
    
        if (i+1) % 500 == 0: plt.show()
        else: plt.close("all")
            
save_gif_PIL("nn.gif", files, fps=20, loop=0)
````

## PINN

> Finally, we add the underlying differential equation ("physics loss") to the loss function. 

The physics loss aims to ensure that the learned solution is consistent with the underlying differential equation. This is done by penalising the residual of the differential equation over a set of locations sampled from the domain.

Here we evaluate the physics loss at 30 points uniformly spaced over the problem domain $([0,1])$. We can calculate the derivatives of the network solution with respect to its input variable at these points using `pytorch`'s autodifferentiation features, and can then easily compute the residual of the differential equation using these quantities.

````python
x_physics = torch.linspace(0,1,30).view(-1,1).requires_grad_(True)# sample locations over the problem domain
mu, k = 2*d, w0**2

torch.manual_seed(123)
model = FCN(1,1,32,3)
optimizer = torch.optim.Adam(model.parameters(),lr=1e-4)
files = []
for i in range(20000):
    optimizer.zero_grad()
    
    # compute the "data loss"
    yh = model(x_data)
    loss1 = torch.mean((yh-y_data)**2)# use mean squared error
    
    # compute the "physics loss"
    yhp = model(x_physics)
    dx  = torch.autograd.grad(yhp, x_physics, torch.ones_like(yhp), create_graph=True)[0]# computes dy/dx
    dx2 = torch.autograd.grad(dx,  x_physics, torch.ones_like(dx),  create_graph=True)[0]# computes d^2y/dx^2
    physics = dx2 + mu*dx + k*yhp# computes the residual of the 1D harmonic oscillator differential equation
    loss2 = (1e-4)*torch.mean(physics**2)
    
    # backpropagate joint loss
    loss = loss1 + loss2# add two loss terms together
    loss.backward()
    optimizer.step()
    
    
    # plot the result as training progresses
    if (i+1) % 150 == 0: 
        
        yh = model(x).detach()
        xp = x_physics.detach()
        
        plot_result(x,y,x_data,y_data,yh,xp)
        
        file = "plots/pinn_%.8i.png"%(i+1)
        plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
        
        if (i+1) % 6000 == 0: plt.show()
        else: plt.close("all")
            
save_gif_PIL("pinn.gif", files, fps=20, loop=0)
````

# Extension: solving an inverse problem

Let's say instead we were only given x, y values and wanted to find mu and k. We could approach this using
a similar architecture (a PINN).

Idea is the physical parameters are added as learnable scalars to the optimiser alongside the network weights. The loss function doesn't need to change because our physics residual already includes mu and k as multiplicative factors so they're differentiable via autograd.



Why not just use the raw data itself: in a perfect world with continuous noiseless values, you could read off dy/dx and d^2y/dx^2 directly and then use least squares regression. However, imagine you have a collection of noisy points - we still need y, dy/dx, d^2y/dx^2 everywhere and differentiation of noisy data is ill-posed. So, by using a PINN we can avoid amplifying errors and provide a smooth (predicted) analytical function the ODE can be evaluated against. 
> Reference: Definitions and examples of inverse and ill-posed problems https://www.researchgate.net/publication/228636348_Definitions_and_examples_of_inverse_and_ill-posed_problems -- details how naturally inverse problems are ill-posed. 

A problem I had was mu and k take too long to converge and that's because the physics loss is weighted too small and thats what produces the gradient signal for mu and k.  

````python
# chore: refactor duplicated code, for now i just commented whats been changed

mu_pred = nn.Parameter(torch.tensor([15.0])) # <- strategy 1: use rough estimates of true values
k_pred = nn.Parameter(torch.tensor([400.0]))

torch.manual_seed(123)
model = FCN(1,1,32,3)

# changed! <- include the new parameters we want to learn
optimizer = torch.optim.Adam(list(model.parameters())+ [mu_pred, k_pred],
                             lr=1e-2)
files = []
for i in range(20000):
    optimizer.zero_grad()
    
    # compute the "data loss"
    yh = model(x_data)
    loss1 = torch.mean((yh-y_data)**2)# use mean squared error
    
    # compute the "physics loss"
    yhp = model(x_physics)
    dx  = torch.autograd.grad(yhp, x_physics, torch.ones_like(yhp), create_graph=True)[0]# computes dy/dx
    dx2 = torch.autograd.grad(dx,  x_physics, torch.ones_like(dx),  create_graph=True)[0]# computes d^2y/dx^2
    physics = dx2 + mu_pred*dx + k_pred*yhp # changed! <- use our mu_pred and k_pred
    # changed! <- physics loss more important for mu and k
    loss2 = (1e-4)*torch.mean(physics**2)
    
    # backpropagate joint loss
    loss = loss1 + loss2 # add two loss terms together
    loss.backward()
    optimizer.step()
    
    
    # plot the result as training progresses
    if (i+1) % 150 == 0: 
        
        yh = model(x).detach()
        xp = x_physics.detach()

        # changed! <- pass predicted mu and k
        plot_result(x, y, x_data, y_data, yh, xp, mu=mu_pred, k=k_pred, mu_true=mu, k_true=k)
        
        file = "plots/inverse_pinn_%.8i.png"%(i+1)
        plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
        
        if (i+1) % 1000 == 0: plt.show()
        else: plt.close("all")
            
save_gif_PIL("inverse_pinn.gif", files, fps=20, loop=0)
````

# What if we started at bad estimates?

In the above example, we show a simple demo of how to change the problem structure to learn paramters part of the model. The things I noticed make a dramatic difference: modify the loss function to weigh more importance of physics loss (consequently lose general shape), learning rate (because our physics loss is used to modify k, mu and its signal is small it can take a long time for our parameters to converge and move - by using a more explosive learning rate our gradient converge faster). But we still had the advantage of starting close to the target solution.

````python
# chore: refactor duplicated code, for now i just commented whats been changed

mu_pred = nn.Parameter(torch.tensor([1.0])) # <- strategy 1: use rough estimates of true values
k_pred = nn.Parameter(torch.tensor([1.0]))

torch.manual_seed(123)
model = FCN(1,1,32,3)

# changed! <- include the new parameters we want to learn
optimizer = torch.optim.Adam(list(model.parameters())+ [mu_pred, k_pred],
                             lr=1e-2)
files = []
for i in range(30000):
    optimizer.zero_grad()
    
    # compute the "data loss"
    yh = model(x_data)
    loss1 = torch.mean((yh-y_data)**2)# use mean squared error
    
    # compute the "physics loss"
    yhp = model(x_physics)
    dx  = torch.autograd.grad(yhp, x_physics, torch.ones_like(yhp), create_graph=True)[0]# computes dy/dx
    dx2 = torch.autograd.grad(dx,  x_physics, torch.ones_like(dx),  create_graph=True)[0]# computes d^2y/dx^2
    physics = dx2 + mu_pred*dx + k_pred*yhp # changed! <- use our mu_pred and k_pred
    # changed! <- physics loss more important for mu and k
    loss2 = (1e-4)*torch.mean(physics**2)
    
    # backpropagate joint loss
    loss = loss1 + loss2 # add two loss terms together
    loss.backward()
    optimizer.step()
    
    
    # plot the result as training progresses
    if (i+1) % 150 == 0: 
        
        yh = model(x).detach()
        xp = x_physics.detach()

        # changed! <- pass predicted mu and k
        plot_result(x, y, x_data, y_data, yh, xp, mu=mu_pred, k=k_pred, mu_true=mu, k_true=k)
        
        file = "plots/inverse_pinn_bad_start_%.8i.png"%(i+1)
        plt.savefig(file, bbox_inches='tight', pad_inches=0.1, dpi=100, facecolor="white")
        files.append(file)
        
        if (i+1) % 1000 == 0: plt.show()
        else: plt.close("all")
            
save_gif_PIL("inverse_pinn_bad_start.gif", files, fps=20, loop=0)
````
